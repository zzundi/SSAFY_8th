```python
n = int(input())

for j in range(n):
    howls = list(input())
    croak = {'c':0, 'r':0, 'o':0, 'a':0}
    num_frog = 0
    for i, howl in enumerate(howls):
        if min(croak.values()) < 0:
            print('#{} {}'.format(j+1, -1))
            break
        
        num_frog = max(num_frog, croak['c'] + croak['r'] + croak['o'] + croak['a'])
        
        if i+1 == len(howls):
            if howl != 'k':
                print('#{} {}'.format(j+1, -1))
                break
            else:
                croak['a'] -= 1
                if any(croak.values()):
                    print('#{} {}'.format(j+1, -1))
                    break
                else:
                    print('#{} {}'.format(j+1, num_frog))
                    break
            
                
        
        if howl == 'c':
            if i == 0 and howl != 'c':
                print('#{} {}'.format(j+1, -1))
                break
            else:
                croak[howl] += 1
                continue
            
        if howl == 'r':
            if croak['c'] > 0:
                croak[howl] += 1
                croak['c'] -= 1
                continue
            else:
                print('#{} {}'.format(j+1, -1))
                break

        if howl == 'o':
            if croak['r'] > 0:
                croak[howl] += 1
                croak['r'] -= 1
                continue
            else:
                print('#{} {}'.format(j+1, -1))
                break
        
        if howl == 'a':
            if croak['o'] > 0:
                croak[howl] += 1
                croak['o'] -= 1
                continue
            else:
                print('#{} {}'.format(j+1, -1))
                break

        if howl == 'k':
            if croak['a'] > 0:
                croak['a'] -= 1
                continue
            else:
                print('#{} {}'.format(j+1, -1))
                break
        
        

```
